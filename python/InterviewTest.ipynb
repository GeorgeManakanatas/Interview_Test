From above we see that many columns have just one value, these will be discarded.\
We also see that there are some NaN values since the count is often smaller than the lines provided by the shape. If more than 20% of the total are missing we set the column aside.\

Based on the explanation file we will focus our attention on 3 aspects of the dataframe
* one realated to the machine's state such as the various statuses, codes, modes etc (all of them have less than 5 distinct values)
* another related to the physical parameters of the machine speed position acceleration etc

```python
# we may want ot reuse this
def percentage_missing_values(column):
    '''
    Gets the percentage of missing values in a column

    Arguments:
        dataframe column

    Returns:
        numpy.float64
    '''
    percentage_missing_values = np.true_divide(column.isna().sum(),
                                               column.shape[0])
    return percentage_missing_values


# lists of columns to be discarded or copied
only_one_value = []
problem_column = []
#
for col in cnc_data.columns:
    if cnc_data[col].nunique() < 2:
        only_one_value.append(col)
    else:
        # if too many values missing, or types are mixed, to problem column
        if percentage_missing_values(cnc_data[col]) > 0.2:
            problem_column.append(col)
        elif cnc_data[col].dtypes != 'float64':
            problem_column.append(col)
#
cnc_data.drop(only_one_value, inplace=True, axis=1)

# Building the problem dataframe
problematic_columns = pd.DataFrame(index=cnc_data.index)
#
for column in problem_column:
    column_to_copy = cnc_data.pop(column)
    problematic_columns.insert(0, column, column_to_copy)
```

At this point we have no column with more than 8% of the data missing.\

The minimum and maximum values of the columns are rather close in terms of scale.

Moving on to making more exploratory plots

```python
def plot_values_with_index(column):
    '''
    Plots the provided column values against it's index.
    X and Y dimensions correspond to the min and max values
    of the index and the column respectively

    Arguments:
        Dataframe column

    Returns:
        ---
    '''
    # setting min and max values for x and y. Y has a 10% margin added
    y_max_value = column.max()*1.1
    y_min_value = column.min()*1.1
    x_max_value = column.index.max()
    x_min_value = column.index.min()
    # creating the plots
    plt.rcParams["figure.figsize"] = (25, 3)
    plt.ylim([y_min_value, y_max_value])
    plt.xlim([x_min_value, x_max_value])
    plt.title(column.name)
    plt.plot(column.index, column)
    plt.show()


def plot_barchart(column):
    '''
    Plots a barchart of the values on the X axis and
    how frequent they are as a percent of total on Y.
    Y axis is boviously from 0 to 100 and the X axis
    has all the discrete values present in the column
    excluding null and NaN

    Arguments:
        Dataframe column

    Returns:
        ---
    '''
    #
    plt.rcParams["figure.figsize"] = (4, 4)
    #
    x = column.value_counts().index
    y = 100*column.value_counts()/column.shape[0]
    #
    x_pos = []
    counter = 0
    for value in x:
        x_pos.append(counter)
        counter += 1
    #
    plt.bar(x_pos, y, )
    plt.ylim([0, 100])
    plt.xlabel("Code")
    plt.ylabel("Timestamps")
    plt.title(column.name)
    plt.xticks(x_pos, x)
    plt.show()


def correlation_heatmap(correlation):
    '''
    Generates correlation heatmap from precalculated correlation matrix

    Arguments:
        Correlation matrix

    Returns:
        ---
    '''
    fig, ax = plt.subplots(figsize=(10, 10))
    sns.heatmap(correlation, cmap="Blues")


# Looking to build 5 lists
position_list = []
velocity_list = []
position_and_velocity = []
sp1_info = []
sp2_info = []
try:
    for col in cnc_data.columns:
        if cnc_data[col].nunique() <= 5:
            plot_barchart(cnc_data[col])
        elif 'actToolBasePos' in col:
            position_list.append(col)
        elif 'ActVelocityMCS' in col:
            velocity_list.append(col)
        elif ('SP1' in col) or ('_module1_Channel_1' in col):
            sp1_info.append(col)
        elif ('SP2' in col) or ('_module1_Channel_2' in col):
            sp2_info.append(col)
        else:
            plot_values_with_index(cnc_data[col])
except Exception as e:
    raise

position_list.append('Alarm_no')
velocity_list.append('Alarm_no')

try:
    correlation_heatmap(cnc_data[sp1_info].corr())
    correlation_heatmap(cnc_data[sp2_info].corr())
    #
    correlation_heatmap(cnc_data[position_list].corr())
    correlation_heatmap(cnc_data[velocity_list].corr())
except Exception as e:
    raise
```
